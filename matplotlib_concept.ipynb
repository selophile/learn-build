```python
import matplotlib.pyplot as plt 
```

2D plot:
It is used to be 2 column.
Both data need to be numerical 
or 
one is numerical and other is categorical

Use Case:
Time series data like stock,covid cases.

```python
price=[1000,21000,35000,48000,56700]
year=[2015,2016,2017,2018,2019]
```

```python
plt.plot(year,price)
plt.show()
```

```python
run=[10000,21000,4539,6482,1283]
batsman=['sachin','rohit','virat','shubham','hardik']
plt.plot(batsman,run,label="batsman run")
plt.xlabel("batsman name")
plt.ylabel("batsman run")
plt.legend()
plt.show()
```

Scatter plot:
  It is used to study 2 columns at a same time 
  numeric vs numeric .

   use cases:
   finding correlation.


```python
import numpy as np
x=np.linspace(-10,10,50)
y=10*+np.random.randint(0,300,50)
plt.scatter(x,y)
plt.xlabel("number of experience")
plt.ylabel("salary")
plt.show()

```

```python
import numpy as np
x=np.linspace(1,10,50)
y=100*+np.random.randint(0,300,50)
plt.scatter(x,y)
plt.xlabel("number of experience")
plt.ylabel("salary")
plt.show()
```

Bar chart:

bivariate analysis: Numerical vs categorical value.


use case: Aggregate analysis of group



```python
childer=[10,20,40,10,30]
color=['red','blue','green','yellow','black']
ax=plt.axes()
plt.bar(color,childer,color='yellow')
ax.set_facecolor('red')
plt.show()
```
