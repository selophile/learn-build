```python
def calculate_grade(score):
    if 300>=score>270:
        return 'A'
    elif 270>score>=255:
        return 'B'
    elif 255>score>=105:
        return 'C'
    elif score<105:
        return 'Fail'
    else:
        return 'Enter correct value'
    
```

```python
merged_data['Grade']=merged_data['Total_Score'].apply(calculate_grade)
merged_data
```

```python
df=merged_data
df
```

```python
filter_data=df[(df['Gender']=='M') & (df['Math']>90)]
filter_data
```
